```python
# ê¸°ë³¸ ë¼ì´ë¸ŒëŸ¬ë¦¬ë¼ íŒŒì´ì¬ ì„¤ì¹˜ì‹œ ìë™ìœ¼ë¡œ ì„¤ì¹˜ë©ë‹ˆë‹¤!
# random ëª¨ë“ˆì€ ëœë¤í•œ ìˆ«ìë¥¼ ìƒì„±í•  ë•Œ ì‚¬ìš©í•©ë‹ˆë‹¤.
import random

# ê²Œì„ì— ì‚¬ìš©ë  ë‹¨ì–´ ëª©ë¡
words = ["apple", "banana", "orange", "grape", "lemon"]

# í–‰ë§¨ ê·¸ë¦¼
hangman_pics = [
    """
     ------
     |    |
     |
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |    |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\
     |   /
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\
     |   / \
     |
    ---""",
]

class HangmanGame:
    def __init__(self):
        self.word = random.choice(words)
        self.guessed_letters = []
        self.wrong_guesses = 0
        self.max_wrong_guesses = len(hangman_pics) - 1
        self.current_state = ["_"] * len(self.word)

    def display(self):
        print(hangman_pics[self.wrong_guesses])
        print("Current word: " + " ".join(self.current_state))
        print("Guessed letters: " + ", ".join(self.guessed_letters))

    def play(self):
        print("í–‰ë§¨ ê²Œì„ì— ì˜¤ì‹  ê²ƒì„ í™˜ì˜í•©ë‹ˆë‹¤!")
        while self.wrong_guesses < self.max_wrong_guesses:
            self.display()
            guess = input("Guess a letter: ").lower()

            if guess in self.guessed_letters:
                print("ì´ë¯¸ ì¶”ì¸¡í•œ ë¬¸ìì…ë‹ˆë‹¤. ë‹¤ì‹œ ì‹œë„í•˜ì„¸ìš”.")
                continue

            self.guessed_letters.append(guess)

            if guess in self.word:
                for i, letter in enumerate(self.word):
                    if letter == guess:
                        self.current_state[i] = guess
                if "_" not in self.current_state:
                    print("ì¶•í•˜í•©ë‹ˆë‹¤! ë‹¹ì‹ ì´ ì´ê²¼ìŠµë‹ˆë‹¤!")
                    print("ë‹¨ì–´: " + self.word)
                    break
            else:
                self.wrong_guesses += 1
                if self.wrong_guesses == self.max_wrong_guesses:
                    self.display()
                    print("ë‹¹ì‹ ì´ ì¡ŒìŠµë‹ˆë‹¤. ì •ë‹µì€ " + self.word + "ì˜€ìŠµë‹ˆë‹¤.")

# ê²Œì„ ì‹¤í–‰
if __name__ == "__main__":
    game = HangmanGame()
    game.play()
```
